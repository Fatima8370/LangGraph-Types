```python
from typing import TypedDict, Dict, List
from langgraph.graph import StateGraph, END
import random
```

```python
class AgentState(TypedDict):
    name : str
    number: List[int]
    counter : int
    
```

```python
def greeting_node(state: AgentState) -> AgentState:
    """This node greets the user"""
    state["name"] = f"Hi there, {state["name"]}"
    state["counter"] = 0
    
    print ("greeting node intializing counter to 0")

    return state

def random_node (state: AgentState) -> AgentState:
    """Generates random number from 1 - 10"""
    state["number"].append(random.randint(1,10))
    state["counter"] +=1
    return state

def should_continue(state: AgentState) -> AgentState:
    """Function To decide what to do next"""
    
    if state["counter"] < 5: # continue looping till 5 times
        print ("Entering Loop", state["counter"])
        return "loop"
    
    else : # exit looping
        return "exit"
    
    
```

```python
graph = StateGraph(AgentState)
graph.add_node("greeting", greeting_node)
graph.add_node("random", random_node)
graph.add_edge("greeting", "random")

graph.add_conditional_edges(
    
    "random",
    should_continue,
    {
        "loop": "random",
        "exit": END
    }
)

graph.set_entry_point("greeting")

app = graph.compile()
```

```python
app.invoke({"name":"Fatima", "number":[],"counter":-1})
```
